### Modeling

```python
X = df.drop('target', axis = 1)
y = df['target']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
## use data: 20% test and 80% train 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

feature_names = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
model = DecisionTreeClassifier(random_state=1, max_depth=3)
model.fit(X_train, y_train)
```

```python
predictions_train = model.predict(X_train)
predictions = model.predict(X_test)
```

```python
print(classification_report(y_train, predictions_train))
print(accuracy_score(y_train, predictions_train))
```

```python
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
from sklearn import tree
# Model representation 
text_representation = tree.export_text(model)
print(text_representation)
```

```python
# Print Decision Tree
fig = plt.figure(figsize=(30,20))
_ = tree.plot_tree(model, 
                   feature_names=X_train.columns,  
                   filled=True)
```

```python
# Feature importance
plt.barh(model.feature_names_in_, model.feature_importances_)## Random Forest
```
